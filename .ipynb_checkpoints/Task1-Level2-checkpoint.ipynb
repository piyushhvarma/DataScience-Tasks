```python
import pandas as pd
df = pd.read_csv("Dataset.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
#City, Country Code Distributions 
print(df['Country Code'].value_counts())
print(df['City'].value_counts())
```

```python
print(df['Cuisines'].value_counts())
```

```python
#Top 10 cuisines
print(df['Cuisines'].str.split(',').explode().value_counts().head(10))
print(df['Cuisines'].str.split(',').explode().value_counts().head(10).plot(kind='bar'))
```

```python
#Top 10 cities 
print(df['City'].value_counts().head(10))
print(df['City'].value_counts().head(10).plot(kind='bar'))
```
