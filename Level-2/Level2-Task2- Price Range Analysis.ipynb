```python
import pandas as pd
df = pd.read_csv("Dataset.csv")
```

```python
#Subtask 1 - Most Common Price Range

most_common_price = df['Price range'].value_counts().idxmax()
counts = df['Price range'].value_counts()
print("Most Common Price : ", most_common_price)
print(counts)
```

```python
#SUbtask 2 - Average Ratings

avg_rating_by_price = df.groupby('Price range')['Aggregate rating'].mean()
print("Average Rating for Each Price Range:")
print(avg_rating_by_price)
```

```python
#SUbtask 3 - COlor rating 
color_by_price = df.groupby('Price range')['Rating color'].apply(lambda x: x.mode()[0])
print("Most Common Color by Price Range:")
print(color_by_price)
```
