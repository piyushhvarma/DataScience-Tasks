```python
import pandas as pd
df = pd.read_csv('Dataset.csv')
```

```python
#Additional features
df['name_length'] = df['Restaurant Name'].str.len()
df['address_length'] = df['Address'].str.len()
```

```python
df['name_length'] = df['Restaurant Name'].astype(str).str.len()
df['address_length'] = df['Address'].astype(str).str.len()
```

```python
df['table_booking_flag'] = df['Has Table booking'].apply(lambda x: 1 if x == "Yes" else 0)
df['online_delivery_flag'] = df['Has Online delivery'].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
df['num_cuisines'] = df['Cuisines'].astype(str).str.split(',').apply(len)
```
