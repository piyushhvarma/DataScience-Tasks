```python
import pandas as pd
df = pd.read_csv('Dataset.csv')
```

```python
#Subtask 1 : Determine the percentage of restaurants that offer table booking and online delivery.

table_booking = df['Has Table booking'].value_counts(normalize=True)*100
print("Table booking Availability (in %)")
print(table_booking)
```

```python
online_delivery = df['Has Online delivery'].value_counts(normalize=True)*100
print("Online Delivery Availability (in %)")
print(online_delivery)
```

```python
#Subtask 2 - Compare the average ratings of restaurants with table booking and those without

ratings = df.groupby('Has Table booking')['Aggregate rating'].mean()
print("Average Rating Comparison")
print(ratings)
```

```python
#Analyze the availability of online delivery among restaurants with different price ranges.

delivery_by_price = df.groupby('Price range')['Has Online delivery'].value_counts(normalize=True).unstack() * 100
print("Online Delivery % by Price Range:")
print(delivery_by_price)
```

```python
#Visualization
df['Has Table booking'].value_counts().plot(kind='bar', title='Table Booking Availability')

```

```python
df['Has Online delivery'].value_counts().plot(kind='bar', title='Online Delivery Availability')
delivery_by_price.plot(kind='bar', figsize=(8,4), title='Online Delivery by Price Range')
```
