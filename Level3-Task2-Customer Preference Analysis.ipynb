```python
import pandas as pd 
df = pd.read_csv('Dataset.csv')
```

```python
# Split cuisines and explode into separate rows
cuisine_ratings = df[['Cuisines','Aggregate rating']].copy()
cuisine_ratings['Cuisines'] = cuisine_ratings['Cuisines'].astype(str).str.split(',')
cuisine_ratings = cuisine_ratings.explode('Cuisines')
cuisine_ratings['Cuisines'] = cuisine_ratings['Cuisines'].str.strip()

# Average rating for each cuisine
avg_rating_by_cuisine = cuisine_ratings.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

print(avg_rating_by_cuisine.head(20))   # top 20 cuisines by rating
```
