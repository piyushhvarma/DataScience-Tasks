```python
import pandas as pd 
df = pd.read_csv('Dataset.csv')
```

```python
# Split cuisines and explode into separate rows
cuisine_ratings = df[['Cuisines','Aggregate rating']].copy()
cuisine_ratings['Cuisines'] = cuisine_ratings['Cuisines'].astype(str).str.split(',')
cuisine_ratings = cuisine_ratings.explode('Cuisines')
cuisine_ratings['Cuisines'] = cuisine_ratings['Cuisines'].str.strip()

# Average rating for each cuisine
avg_rating_by_cuisine = cuisine_ratings.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)

print(avg_rating_by_cuisine.head(20))   # top 20 cuisines by rating
```

```python
# Keep Cuisines + Votes
cuisine_votes = df[['Cuisines','Votes']].copy()
cuisine_votes['Cuisines'] = cuisine_votes['Cuisines'].astype(str).str.split(',')
cuisine_votes = cuisine_votes.explode('Cuisines')
cuisine_votes['Cuisines'] = cuisine_votes['Cuisines'].str.strip()

# Sum of votes for each cuisine
popular_cuisines = cuisine_votes.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False)

print(popular_cuisines.head(20))
```

```python
high_rating_cuisines = avg_rating_by_cuisine[avg_rating_by_cuisine > 4.0]
print("Cuisines With Avg Rating Above 4.0:")
print(high_rating_cuisines)
```
