```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set(style="whitegrid")

df = pd.read_csv('Dataset.csv')
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df['Aggregate rating'], bins=10, kde=True)
plt.title("Distribution of Aggregate Ratings")
plt.show()
```

```python
plt.figure(figsize=(7,4))
df['Aggregate rating'].value_counts().sort_index().plot(kind='bar')
plt.title("Rating Count Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
df_cuis = df.copy()
df_cuis['Cuisines'] = df_cuis['Cuisines'].astype(str).str.split(',')
df_cuis = df_cuis.explode('Cuisines')
df_cuis['Cuisines'] = df_cuis['Cuisines'].str.strip()
```

```python
top_cuisine_ratings = df_cuis.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,4))
top_cuisine_ratings.plot(kind='bar')
plt.title("Top 10 Cuisines by Average Rating")
plt.ylabel("Average Rating")
plt.show()
```
